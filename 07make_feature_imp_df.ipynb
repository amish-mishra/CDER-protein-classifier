```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
import pickle
import seaborn as sns
import random
from IPython.display import display
```

### SME Feature Importance

```python
top_arr = ['HHH', 'EEHEE', 'HEEH', 'EHEE']
num_iterations = 10
save_path = './sme_feature_importances_dataframes'
# Get feature names
raw_df = pandas.read_csv('main_df.csv')
feature_names = raw_df.iloc[:,4:].columns.to_list()

for i, top in enumerate(top_arr):
    print(top)
    all_forest_importances = pandas.DataFrame(index=feature_names)
    for num in range(num_iterations):
        clf_file = f'classifiers/{top}_sme_rf_clf_gs_{num}.pickle'
        with (open(clf_file, "rb")) as openfile:
            clf = pickle.load(openfile)
        forest = clf.best_estimator_
        importances = forest.feature_importances_
        importances_df = pandas.DataFrame(importances, index=feature_names, columns=[f'Iteration {num}'])
        all_forest_importances = pandas.concat([all_forest_importances, importances_df], axis=1)
    display(all_forest_importances)
    # Save data frame
    all_forest_importances.to_csv(f'{save_path}/{top}_sme_feature_importances_df.csv', index=True)
```

### SME Feature Importance

```python
top_arr = ['HHH', 'EEHEE', 'HEEH', 'EHEE']
num_iterations = 10
save_path = './sme_feature_importances_dataframes'
# Get feature names
raw_df = pandas.read_csv('main_df.csv')
feature_names = raw_df.iloc[:,4:].columns.to_list()

for i, top in enumerate(top_arr):
    print(top)
    all_forest_importances = pandas.DataFrame(index=feature_names)
    for num in range(num_iterations):
        clf_file = f'classifiers/{top}_sme_rf_clf_gs_{num}.pickle'
        with (open(clf_file, "rb")) as openfile:
            clf = pickle.load(openfile)
        forest = clf.best_estimator_
        importances = forest.feature_importances_
        importances_df = pandas.DataFrame(importances, index=feature_names, columns=[f'Iteration {num}'])
        all_forest_importances = pandas.concat([all_forest_importances, importances_df], axis=1)
    display(all_forest_importances)
    # Save data frame
    all_forest_importances.to_csv(f'{save_path}/{top}_sme_feature_importances_df.csv', index=True)
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,12))
plt.suptitle('Mean of feature importances over 10 iterations (with 1 std. dev. bars)')
    
for i, top in enumerate(top_arr):
    # Find the mean of the feature importances across all iterations
    features_df = pandas.read_csv(f'{save_path}/{top}_sme_feature_importances_df.csv', index_col=0)
    avg_feature_imp_df = features_df.mean(axis=1)
    std_imp = features_df.std(axis=1)
    
    # Choose the largest
    largest_importances = avg_feature_imp_df.nlargest(n=20)

    # Plot
    j = 1 if i > 1 else 0
    i = i%2
    largest_importances.plot.barh(ax=axs[i][j], xerr=std_imp, error_kw={'linewidth': 1, 'capsize': 2, 'capthick': 1})
    axs[i][j].set_title("Feature importances for "+top)
    axs[i][j].set_xlabel("Mean decrease in impurity")
    fig.tight_layout()
    
# fig.savefig('./figures/average_feature_importances.png')
```

### SME Feature Importance of no train/test split model

```python
top_arr = ['HHH', 'EEHEE', 'HEEH', 'EHEE']
save_path = './sme_feature_importances_dataframes'

# Get feature names
raw_df = pandas.read_csv('main_df.csv')
feature_names = raw_df.iloc[:,4:].columns.to_list()

for i, top in enumerate(top_arr):
    all_forest_importances = pandas.DataFrame(index=feature_names)
    clf_file = f'classifiers/{top}_sme_rf_clf_gs_no_train_test_split.pickle'
    with (open(clf_file, "rb")) as openfile:
        clf = pickle.load(openfile)
    forest = clf.best_estimator_
    importances = forest.feature_importances_
    importances_df = pandas.DataFrame(importances, index=feature_names, columns=['feature_imp'])
    all_forest_importances = pandas.concat([all_forest_importances, importances_df], axis=1)
#     display(all_forest_importances)
    # Save data frame
    all_forest_importances.to_csv(f'{save_path}/{top}_sme_feature_importances_no_train_test_split_df.csv', index=True)
```

## Plot the most important features for each topology for the no train/test split model

```python
fig, axs = plt.subplots(2,2, figsize=(12,8))
plt.suptitle('SME feature importance of no train/test split model')
    
for i, top in enumerate(top_arr):
    # Find the mean of the feature importances across all iterations
    features_df = pandas.read_csv(f'{save_path}/{top}_sme_feature_importances_no_train_test_split_df.csv', index_col=0)
    
    # Choose the largest
    largest_importances = features_df.nlargest(n=10, columns=['feature_imp'])

    # Plot
    j = 1 if i > 1 else 0
    i = i%2
    largest_importances.plot.barh(ax=axs[i][j], error_kw={'linewidth': 1, 'capsize': 2, 'capthick': 1})
    axs[i][j].set_title("Feature importances for "+top)
    axs[i][j].set_xlabel("Mean decrease in impurity")
    fig.tight_layout()
    
fig.savefig('./figures/sme_feature_importances_no_train_test_split.svg')
```

### CDER coordinates feature importance

```python
top_arr = ['HHH', 'EEHEE', 'HEEH', 'EHEE']
save_path = './cder_feature_importances_dataframes'


for i, top in enumerate(top_arr):
    # Get feature names
    raw_df = pandas.read_csv(f'features_dataframes/sme_cder_train_X_{top}_no_train_test_split.csv')
    feature_names = raw_df.filter(regex=r'^H_').columns
    all_forest_importances = pandas.DataFrame(index=feature_names)
    clf_file = f'classifiers/cder_{top}_rf_clf_gs_no_train_test_split.pickle'
    with (open(clf_file, "rb")) as openfile:
        clf = pickle.load(openfile)
    forest = clf.best_estimator_
    importances = forest.feature_importances_
    importances_df = pandas.DataFrame(importances, index=feature_names, columns=['feature_imp'])
    all_forest_importances = pandas.concat([all_forest_importances, importances_df], axis=1)
#     display(all_forest_importances)
    # Save data frame
    all_forest_importances.to_csv(f'{save_path}/{top}_cder_feature_importances_no_train_test_split_df.csv', index=True)
```

## Plot the most important features for each topology for the no train/test split model

```python
fig, axs = plt.subplots(2,2, figsize=(8,8))
plt.suptitle('CDER feature importance of no train/test split model')
    
for i, top in enumerate(top_arr):
    features_df = pandas.read_csv(f'{save_path}/{top}_cder_feature_importances_no_train_test_split_df.csv', index_col=0)
    
    # Choose the largest
    largest_importances = features_df.nlargest(n=len(features_df), columns=['feature_imp'])
    # Plot
    j = 1 if i > 1 else 0
    i = i%2
    largest_importances.plot.barh(ax=axs[i][j], legend=False)
    # Invert the x and y axes
    axs[i][j].invert_xaxis()
    axs[i][j].invert_yaxis()
    axs[i][j].set_title("Feature importances for "+top)
    axs[i][j].set_xlabel("Mean decrease in impurity")
    fig.tight_layout()
    
fig.savefig('./figures/cder_feature_importances_no_train_test_split.svg')
```
